```python
def create_time_data(dataset):
    X = []
    Y = []
    for i in range(len(dataset)-81):
        a = dataset[i:(i+80),0]
        X.append(a)
        Y.append(dataset[i + 80,0])
    return np.array(X), np.array(Y)
```

```python
X_train,Y_train = create_time_data(df_train)
```

```python
X_test,Y_test = create_time_data(df_test)
```

```python
X_train.shape
```

Reshaping X_train and X_test for LSTM model :

```python
X_train = X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(80,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
history1 = model.fit(X_train,Y_train,validation_data=(X_test,Y_test),epochs=100,batch_size=64,verbose=1)
```

2) Finding number of hidden layers and parameters:

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
import tensorflow as tf
```

```python
from keras.layers import Dense, Activation, Embedding, Flatten, LeakyReLU, BatchNormalization, Dropout
from keras.activations import relu, sigmoid
```

```python
def create_model(layers, activation):
    model = Sequential()
    for i, nodes in enumerate(layers):
        if i==0:      # For first layer
            model.add(LSTM(nodes,return_sequences=True,input_shape=(80,1)))
            model.add(Activation(activation))
        else:
            model.add(LSTM(nodes,return_sequences=True))
            model.add(Activation(activation))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = KerasClassifier(build_fn=create_model, verbose=0)
```

```python
layers = [[20], [50, 20], [50, 50, 50]]
activations = ['sigmoid', 'relu']
param_grid = dict(layers=layers,activation=activations,batch_size = [64,128], epochs=[100])
grid = GridSearchCV(estimator=model, param_grid=param_grid,cv=2)
```

```python
grid_result = grid.fit(X_train, Y_train)
```

```python
grid_result.best_params_
```

```python
grid_df = pd.DataFrame(grid.cv_results_)
```

```python
grid_df[['param_layers','rank_test_score','params']]
```
